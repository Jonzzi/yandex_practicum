### Примечание:
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

```python
# Считаем данные по пользователям
calc_calls = calls.pivot_table(index=['user_id','month'], values='oper_duration', aggfunc=['count','sum']).reset_index()
calc_calls.columns = ['user_id', 'month', 'calls_per_month', 'duration_per_month']
display(calc_calls.head(15))

users_data = users.merge(calc_calls, on='user_id', how='left')
#users_data.info()
users_data = users_data.fillna(0)
#users_data.info()
#print(users_data.describe())

calc_inet = internet.pivot_table(index=['user_id','month'], values='mb_used', aggfunc='sum').reset_index()
calc_inet['mb_used'] = calc_inet['mb_used'] / 1024
calc_inet['mb_used'] = calc_inet['mb_used'].apply(np.ceil)
calc_inet.rename(columns={'mb_used': 'gb_used'}, inplace=True)
users_data = users_data.fillna(0)
display(calc_inet.head(15))
#users_data.info()

users_data.columns = ['user_id', 'age', 'churn_date', 'city', 'first_name', 'last_name', 'reg_date', 'tariff', 'month', 'calls_per_month', 'duration_per_month']
users_data = users_data.merge(calc_inet, on=['user_id', 'month'], how='left')
users_data = users_data.fillna(0)

calc_msg = messages.pivot_table(index=['user_id','month'], values='id', aggfunc='count').reset_index()
display(calc_msg.head(15))
#users_data.info()

users_data = users_data.merge(calc_msg, on=['user_id', 'month'], how='left')
users_data.rename(columns={'id': 'msg_per_month'}, inplace=True)
users_data = users_data.fillna(0)

tariffs.rename(columns={'tariff_name': 'tariff'}, inplace=True)
users_data = users_data.merge(tariffs, on='tariff', how='left')
users_data = users_data.fillna(0)
#users_data.info()

users_data['rub_monthly_income'] = (users_data['rub_monthly_fee'] +
                                    (users_data['gb_used'] - users_data['gb_per_month_included']) * users_data['rub_per_gb'] * (users_data['gb_used'] > users_data['gb_per_month_included']) +
                                    (users_data['duration_per_month'] - users_data['minutes_included']) * users_data['rub_per_minute'] * (users_data['duration_per_month'] > users_data['minutes_included']) +
                                    (users_data['msg_per_month'] - users_data['messages_included']) * users_data['rub_per_message'] * (users_data['msg_per_month'] > users_data['messages_included'])
                                   )
```

* средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
* средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

Пороговое значение alpha задайте самостоятельно.

## Проверка первой гипотезы
Гипотезы формулируем так:
* нулевая гипотеза "Средняя выручка по пользователям разных тарифов НЕ различаются",
* альтернативная гипотеза "Средняя выручка по пользователям разных тарифов различаются"

```python
mn_u = np.ceil(users_data.query('tariff=="ultra"')['rub_monthly_income'].mean())
print('Средняя выручка по тарифу ultra:', mn_u)
mn_s =  np.ceil(users_data.query('tariff=="smart"')['rub_monthly_income'].mean())
print('Средняя выручка по тарифу smart:', mn_s)
results = st.ttest_ind (users_data.query('tariff=="ultra"')['rub_monthly_income'], users_data.query('tariff=="smart"')['rub_monthly_income'], equal_var=False)
print('Используем метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей:')
print('p-значение: ', results.pvalue)
alpha = .01
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу. Средние выручки различаются.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние выручки НЕ различаются")
    
# Количество пользователей с тарифом ultra: 149
# Количество пользователей с тарифом smart: 351    
print ('Оценочная общая выручка за месяц по тарифу ultra (руб.):', ul * mn_u)
print ('Оценочная общая выручка за месяц по тарифу smart (руб.):', sm * mn_s)
```

### Выводы:
1. При такой разнице средних выручек нулевая гипотеза не подтвердится даже при alpha = 0.01. Критерий брал малый, т.к. изначально было понятно, что схожести мало.

## Проверка второй гипотезы
Гипотезы формулируем так:
* нулевая гипотеза "Средняя выручка по пользователям Москвы и других регионов НЕ различаются", 
* альтернативная гипотеза "Средняя выручка по пользователям Москвы и других регионов различаются"

```python
print('Средняя выручка пользователей из Москвы:', np.ceil(users_data.query('city=="Москва"')['rub_monthly_income'].mean()))
print('Средняя выручка пользователей из других регионов:', np.ceil(users_data.query('city!="Москва"')['rub_monthly_income'].mean()))
results = st.ttest_ind (users_data.query('city=="Москва"')['rub_monthly_income'], users_data.query('city!="Москва"')['rub_monthly_income'], equal_var=True)
print('Используем метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей:')
print('p-значение: ', results.pvalue)
alpha = .05
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу. Средние выручки различаются.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние выручки НЕ различаются")
```

### Выводы:
2. Средние действительно почти одинаковы, нулевая гипотеза подтвердилась даже при alpha = 0.05. Критерий брал обычный, т.к. никаких предпосылок для его уменьшения не было.
